### Task 1
**<font>[10 points]Let $\mathbf{X}\in \mathbb{R}^{p\times N}, p \lt N$ be a centered matrix of $p$ $N$-dimensional data samples with the full-size SVD $\mathbf{X}=\mathbf{U}\mathbf{\Sigma}\mathbf{V}^T$. Assume that the singular values are sorted in a descending manner, i.e. $\sigma_{1,1} \geq ... \geq \sigma_{p,p}$.</font>**

#### Task 1.1
**<font>
Provide a normalized vector $\hat{\mathbf{s}}\in\mathbb{R}^p$, such that
$$
\hat{\mathbf{s}} = \arg \max_{\mathbf{s}\; \text{s.t.}\; \|\mathbf{s}\|=1} \mathbf{s}^T\mathbf{\Sigma\Sigma}^T\mathbf{s} \; .
$$
</font>**

Suppose $\mathbf{s}$ = 
${\begin{pmatrix}
s_1 \quad s_2 \quad \dots \quad s_p
\end{pmatrix}}^T$,
$\mathbf{\Sigma}$ = 
${\begin{pmatrix}
\sigma_1 \quad 0 \quad \dots \quad 0
\\
0 \quad \sigma_2 \quad \dots \quad 0
\\
0 \quad \dots \quad \ddots \quad 0
\\
0 \quad \ 0 \quad \dots \quad \sigma_p
\end{pmatrix}}^T$

\begin{equation*}
    \begin{split}
        & so, \ \mathbf{s}^T\mathbf{\Sigma}\mathbf{\Sigma}^T\mathbf{s} = 
        \begin{pmatrix}
        s_1 \quad s_2 \quad \dots \quad s_p
        \end{pmatrix}
        \begin{pmatrix}
        \sigma_1 \quad 0 \quad \dots \quad 0
        \\
        0 \quad \sigma_2 \quad \dots \quad 0
        \\
        0 \quad \dots \quad \ddots \quad 0
        \\
        0 \quad \ 0 \quad \dots \quad \sigma_p
        \end{pmatrix}
        \begin{pmatrix}
        \sigma_1 \quad 0 \quad \dots \quad 0
        \\
        0 \quad \sigma_2 \quad \dots \quad 0
        \\
        0 \quad \dots \quad \ddots \quad 0
        \\
        0 \quad \ 0 \quad \dots \quad \sigma_p
        \end{pmatrix}
        \begin{pmatrix}
        s_1
        \\
        s_2
        \\
        \vdots
        \\
        s_p
        \end{pmatrix}
         = \sum_{i=1}^{p}{s_i}^2{\sigma_i}^2
    \end{split}
\end{equation*}

if $\|\mathbf{s}\|=1$, 
$$
\max_{\Vert \mathbf{s}\Vert=1} \mathbf{s}^{T}\mathbf{\Sigma \Sigma}^{T}\mathbf{s} = \max_{\Vert \mathbf{s} \Vert=1} \sum_{i=1}^{p}{\sigma_i}^2{s_i}^2 = {\sigma_1}^2
$$
Because the singular values are sorted in a descending manner, $\sigma_{1,1} \geq ... \geq \sigma_{p,p}>0$, so the maximal value for 
$s={\begin{pmatrix}
1 \quad 0 \quad \dots \quad 0
\end{pmatrix}}^T$ 
and 
${\begin{pmatrix}
-1 \quad 0 \quad \dots \quad 0
\end{pmatrix}}^T$.
Therefore,
$\tilde{\mathbf{s}}\in \{ {
\begin{pmatrix}
1, \quad 0, \quad \dots \quad 0
\end{pmatrix}}^T, 
{\begin{pmatrix}
-1, \quad 0, \quad \dots \quad 0
\end{pmatrix}}^T\}$

#### Task 1.2
**<font>
Show that the emperical variance of the inner products of the columns of $\mathbf{X}$ with a normalized vector $\mathbf{a}$,
</font>**
$$
\frac{1}{N}\sum_{i=1}^N(\mathbf{a}^T\mathbf{x}_i)^2 = \frac{1}{N}\mathbf{a}^T\mathbf{XX}^T\mathbf{a} \; ,
$$
**<font>
is maximized when $\mathbf{a}$ is set to the first column of $\mathbf{U}$, i.e. $\mathbf{a}=\mathbf{u}_1$ (note that $\| \mathbf{a} \| = 1$).
</font>**

**<font>
Hint: Write $\mathbf{a}$ as a linear combination of the columns of $\mathbf{U}$. Verify that such a representation does not affect the norm constraint.
</font>**

According to $\mathbf{X}=\mathbf{U}\mathbf{\Sigma}\mathbf{V}^T$, $\mathbf{U}^T\mathbf{U}=\mathbf{I}_N$, so

$$
\begin{align*}
\frac{1}{N}\sum_{i=1}^N(\mathbf{a}^T\mathbf{x}_i)^2 
&= \frac{1}{N}\mathbf{a}^T\mathbf{X}\mathbf{X}^T\mathbf{a} \\
&= \frac{1}{N}\mathbf{a}^T\mathbf{U}\mathbf{\Sigma}\mathbf{V}^T\mathbf{V}\mathbf{\Sigma}^T\mathbf{U}^T\mathbf{a}  \\
&= \frac{1}{N}\mathbf{a}^T\mathbf{U}\mathbf{\Sigma}\mathbf{\Sigma}^T\mathbf{U}^T\mathbf{a} \\
&= \frac{1}{N}\mathbf{a}^T\mathbf{U}\mathbf{\Sigma}^2\mathbf{U}^T\mathbf{a} \\
\end{align*}
$$
Because $\mathbf{a}$ is a linear combination of the columns of $\mathbf{U}$, suppose $\mathbf{a}=\mathbf{Up}$, where $\mathbf{p}$ is a constant. Thus, we can get continuously

\begin{align*}
\frac{1}{N}\mathbf{a}^T\mathbf{U}\mathbf{\Sigma}^2\mathbf{U}^T\mathbf{a}
&= \frac{1}{N}\mathbf{p}^T\mathbf{U}^T\mathbf{U}\mathbf{\Sigma}^2\mathbf{U}^T\mathbf{U}\mathbf{p} \\
&= \frac{1}{N}\mathbf{p}^T\mathbf{\Sigma}^2 \mathbf{p}\\
&= \frac{1}{N}\sum_{i=1}^{N}{p_i}^2{\sigma_i}^2
\end{align*}
Since $\| \mathbf{a} \| = 1$,
$$
\|\mathbf{a}\|=\|\mathbf{Up}\|=\sqrt{\mathbf{p}^T\mathbf{U}^T\mathbf{U}\mathbf{p}}=\sqrt{\mathbf{p}^T\mathbf{p}}=\|\mathbf{p}\| = 1
$$
From previous task, $\sigma_{1,1} \geq ... \geq \sigma_{p,p}>0$, implies
$$
\max_{{\| \mathbf{p}\|=1}} \frac{1}{N}\mathbf{p}^{T}\mathbf{\Sigma}\mathbf{\Sigma}^{T}\mathbf{p} 
= \max_{\| \mathbf{p}\|=1} \frac{1}{N}\sum_{i=1}^{N}{p_i}^2 {\sigma_i}^2
= \frac{1}{N}{\sigma_1}^2
$$
and the maximal value when $\mathbf{p}\in \{ {\begin{pmatrix}1,& 0,& \dots,& 0\end{pmatrix}}^T , {\begin{pmatrix}-1,& 0,& \dots,& 0\end{pmatrix}}^T\}$. 
In addition, $\|\mathbf{a}\|=\|\mathbf{Up}\|=\|\mathbf{p}\| = 1$ \
So 
$$
\mathbf{a}=\mathbf{Up} =\mathbf{U} \begin{pmatrix}\pm 1,& 0,& \dots,& 0\end{pmatrix}^T =\pm \mathbf{u}_1
$$
which means that $\mathbf{a}$ is set to the first column of $\mathbf{U}$

The Yale Face Database B consists of single light source images of $10$ subjects, each seen in different poses and illumination conditions. #### Task 2.1
**<font>Write a function that takes as an input matrix $\mathbf{T}$ of vectorized images from subset $0$. The output of this function are the $20$ first singular vectors $\mathbf{U}[:, 1]\,,\;...\,,\;\mathbf{U}[:, 20]$. Display the first $3$ vectors as images, i.e., reshape them to size $50 \times 50$ and display them.</font>**

```python
import os
import sys
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt

# please place the Jupyter Notebook(Assignment 2) file and the yaleBfaces folder in the same root directory
root_directory = sys.path[0]
directory = 'yaleBfaces'
folder = root_directory+'/'+directory
subset_id = [i for i in range(5)]
subject_id = [i for i in range(1, 11)]
num_vec = 20  # U[:,1]-U[:,20]

img = Image.open(folder+'/subset0/person01_01.png')
h = img.size[0]  # height
w = img.size[1]  # width

# set dataset
def dataset():
    T = dict()
    label = dict()  # Do not initialize outside the function!
    
    for i in subset_id:  # read each subset 0-4
        subset = folder +'/' + 'subset' + str(i)
        img_list = os.listdir(subset)  # return image names list
        num_img = len([x for x in img_list])             
        T[i] = np.zeros((h*w, num_img))  # (2500, 70), first subset:70
        label[i] = np.zeros(num_img)  # label: subject
        j = 0
        
        for fp in img_list:  # fp: image name
            
            img = os.path.join(subset, fp)  
            T[i][:,j] = np.float64(Image.open(img)).flatten()/255  # img->array
            label[i][j] = int(fp[6:8])  # person01->01
            j += 1
            
    return T, label

# returns 20 first singular vectors U[:; 1]--U[:; 20]
def sin_vec(T):
    
    mu = np.mean(T, axis=1)
    T_c = T - np.expand_dims(mu, axis=1)
    # Singular Value Decomposition
    [U, Sigma, VT] = np.linalg.svd(T_c, full_matrices=False)
   
    return U[:,0:num_vec]

# data_set(): ({0/1/2/3/4:array},{0/1/2/3/4:array([labels])})
T = dataset()[0][0]
label = dataset()[1][0]
U = sin_vec(T)

fig = plt.figure(figsize=(10, 10))  # width, height in inches

for idx in range(3):
    
    ax = fig.add_subplot(1, 3, idx+1, xticks=[], yticks=[])  # without any ticks
    ax.imshow(U[:,idx].reshape(h,w))
    ax.set_title(f'$U[0,{idx}]$')  
    fig.tight_layout()  # adjust subplot params->fits in to the figure area

plt.show()
```

#### Task 2.2
**<font>Write a function that takes as an input the training set $\mathbf{T}$ (a matrix composed of vectorized pictures from subset $0$), a vector containing the labels of the training set (i.e., if the $i$-th sample belongs to class $j$, the $i$-th entry of the labels vector is $j$. In this exercise $j$ is an integer between $1$ and $10$), the test samples $S$ (a matrix composed of vectorized pictures from subsets $1$-$4$) and the corresponding labels (in a separate vector), the $20$ singular vectors from the first step, and the parameter $k$ that denotes how many of the PCs are used. Use the Euclidean distance to classify each sample image based on its three nearest neighbors (this is done by comparing the test samples with the training samples in the reduced space.). As an output give the fraction of images from $S$ that were misclassified, i.e., the error rate. Repeat this for subsets $1$ through $4$ and for $k=1\,,\;...\,,\;20$. Plot the error rate for each subset.
</font>**

```python
from collections import Counter

k = num_vec
num_neighbor = 3
T_label = dataset()[1][0]

class KNeighborsClassifier:
    def __init__(self, x_train, y_train, num_neighbor):
        self.x_train = x_train
        self.y_train = y_train
        self.n_neighbor = num_neighbor

    def predict(self, x_test):  
        dist = [np.linalg.norm(x_test - i) for i in self.x_train]  # calculate distances
        idx_nearest = np.argsort(dist)  # ascending order, return index
        nearst_label = [self.y_train[i] for i in idx_nearest[:self.n_neighbor]]  # the first three labels [1,1,5]
        count_pairs = Counter(nearst_label)  # ({key(label):value,:,:}).{1.0:3}/{1.0:2,5.0:1}/{1.0:1,5.0:1,10.0:1}
        sorted_count = count_pairs.most_common(self.n_neighbor)  # [(1.0,3)],[(5.0,1),(1.0,2)],[(1.0,1),(5.0,1),(10.0,1)]
        max_count = sorted_count[0][0]
        
        return max_count
    
    # calculate accuracy
    def score(self, x_test, y_test):
        pred_list = []
        
        for x, y in zip(x_test, y_test):  # elements into tuples, return the list composed of these tuples
            pred = self.predict(x)
            pred_list.append(pred)
            
        num_correct = np.sum(pred_list==y_test)
        
        return num_correct / len(x_test)
```

```python
# calculate error rate for each subset
def error_subset(i):
    err = dict()    
    S = dataset()[0][i]  # test: subset1-4
    S_label = dataset()[1][i]
    err = np.zeros(k)
    j=0
    
    for m in range(1, k+1):
        U1 = U[:, 0:m]  # use U to reduce dimensions
        T1 = np.dot(U1.T, T)  # (m,2500)(2500,70)=(m,70) train
        S1 = np.dot(U1.T, S)  # (m,2500)(2500,?)=(m,?) test
            
        # train KNN
        KNN = KNeighborsClassifier(T1.T, T_label, num_neighbor)  
        # test->error
        err[j] = 1 - KNN.score(S1.T, S_label)
    
        j += 1
        
    return err

# plot error rate for each subset
for i in subset_id[1:]:
    plt.plot(np.linspace(1,k,20), error_subset(i), label='subset '+str(i))
    

plt.title('Error rate for subsets 1-4')
plt.xlabel('k')
plt.ylabel('Error rate')
plt.xticks(range(1,k+1,1))
plt.yticks(np.linspace(0.1,1.0,10))
plt.grid()
plt.legend(loc='lower left', fontsize=8)
plt.show()
```

Repeat the above experiment without using the first three singular vectors, i.e., use $k=1\,,\;...\,,\;17$ singular vectors starting from the $4$-th. Plot the error rate as before. How do you explain the difference in recognition rate?
</font>**

```python
k_begin = 4  #  singular vectors starting from the 4-th
U_new = U[:, k_begin-1:]

def error_new_subset(i):
    err = dict()
    S = dataset()[0][i]  # test: subset1-4
    S_label = dataset()[1][i]
    err = np.zeros(k-k_begin+1)
    j=0
    
    for m in range(1, k-k_begin+2):
        U1 = U_new[:, 0:m]  # use U to reduce dimensions
        T1 = np.dot(U1.T, T)  # (m,2500)(2500,70)=(m,70) train
        S1 = np.dot(U1.T, S)  # (m,2500)(2500,?)=(m,?) test
        
        # train KNN
        KNN = KNeighborsClassifier(T1.T, T_label, num_neighbor)  
        # test->error
        err[j] = 1 - KNN.score(S1.T, S_label)
    
        j += 1
           
    return err

# plot error rate for each subset
fig, ax = plt.subplots(4, 1, figsize=(10,10))

for i, idx in enumerate(subset_id[1:]):
    ax[i].plot(range(1,k+1), error_subset(idx), label='No singular vectors of $U_k$ removed')
    ax[i].plot(range(1,k-k_begin+2), error_new_subset(idx), label='First three singular vectors of $U_k$ removed')
#    ax[i].plot(range(k_begin,k+1), error_new_subset()[idx], label='$U[:,4:21]$')
    ax[i].set_xticks(range(1,k+1,1))
    ax[i].set_title(f'Error rate for each subset: {idx}')
    ax[i].set_xlabel('k')
    ax[i].set_ylabel('Error rate')
    ax[i].grid()
    ax[i].legend()
    plt.tight_layout()  # adjust subplot params->fits in to the figure area
    
plt.show()
```

From subset 1, we can see the two curves are very similar, namely, there is little difference in the error rate between removing the first three columns and not removing the first three columns. In other words, under the relatively strong light source, the role of the first three columns of $U_k$ is not significant. 

And in subsets 2 and 3, there is a large deviation between the two curves. Removing the first three columns of $U_k$ has a relatively better performance. It can greatly reduce the error rate. Although the first three singular vectors contain the most face features, they also contain a lot of negative information about the black parts of the photo. Compared with subset 1, removing the first three cols of $U_k$ has caused a better error rate. It can improve the recognition accuracy under relative dark light.

However, from subset 4, obviously, removing the first three columns of $U_k$ has a relatively worse performance. In this situation, all the singular vectors contain the negative information about the black parts of the photo. removing the first three cols of $U_k$ has caused that the most face features lacked. It leads to a higher error rate.

From overall subsets, the worse the light source, the worse the classification effect. In subset 4, the error rate surprisingly fluctuates about 90%.

The singular vector of principle component is ambiguous and is difficult to be evaluated in an accurate manner. The principal components extracted from different data sets will have different effects on classification.
