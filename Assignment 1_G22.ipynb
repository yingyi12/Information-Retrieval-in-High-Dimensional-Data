**<font>[2 Points] Let $\mathcal{C}_{d}=\left\{\mathrm{x} \in \mathbb{R}^{p} \mid\|\mathrm{x}\|_{\infty} \leq \frac{d}{2}\right\}$ denote the $p$-dimensional hypercube of edge length $d$, centered at the origin. </font>**

### Task 1.1
**<font>Assume $X$ to be uniformly distributed in $\mathcal{C}_{1}$. Determine $d$ in dependence of $p$ and $q \in[0,1]$, such that </font>**

$$
\operatorname{Pr}\left(X \in \mathcal{C}_{d}\right)=q
$$

**<font>holds.</font>**

Since $X$ to be uniformly distributed in $\mathcal{C}_{1}$, and $\mathcal{C}_{1}=\left\{\mathrm{x} \in \mathbb{R}^{p} \mid\|\mathrm{x}\|_{\infty} \leq \frac{1}{2}\right\}$, therefore: 
$$ 
f(x) =\begin{cases}
1 & \quad \text{for } x\in \mathcal{C}_1 \\
0 & \quad \text{otherwise} 
\end{cases}
$$
assume $d \geq 1$, which means $\mathcal{C}_1 \subseteq \mathcal{C}_d$ :
 $$
 \begin{align*}
 & \because \mathcal{C}_1 \subseteq \mathcal{C}_d\\
 & \because X \in \mathcal{C}_{1}\\
 & \therefore X \in \mathcal{C}_{d}\\
 & \therefore \operatorname{Pr}(X \in \mathcal{C}_d)= 1\\
 \end{align*}
 $$

assume $d < 1$:

 For each dimension,
 $$
 \operatorname{Pr}(\|\mathrm{x}\|^i \leq \frac{d}{2})= \dfrac{\frac{d}{2}}{\frac{1}{2}}= d
 $$
 where $\|\mathrm{x}\|^i$ denotes the elements in i-th dimension.\
 Hence, since components of $X^p$ are independent, the probability of $X$ belongs in $C_d$ is
 $$
 \begin{align*}
 \operatorname{Pr}(X \in \mathcal{C}_d)& = (\operatorname{Pr}(\|\mathrm{x}\|^i \leq \frac{d}{2}))^p
 =d^p
 =q
 \end{align*}
 $$

In conclusion, it follows for $q <1$ and $d < 1$,
 $$
 \begin{align*}
 & \operatorname{Pr}(X\in\mathcal{C}_d) = q \\
 & \Rightarrow \, d= q^\frac{1}{p}
 \end{align*}
 $$

### Task 2.1
**<font>Sample $100$ uniformly distributed random vectors from the box $[-1, 1]^{d}$ for $d = 2$.</font>**

```python
import numpy as np
import matplotlib.pyplot as plt

# set the seed of the random number generator to a fixed value, the results will be reproducible.
np.random.seed(1)


def samples(n,d):
    vec =np.random.uniform(-1, 1, (n, d))
    return vec

# generate 100 uniformly distributed random vectors
vec = samples(100, 2)
# visualize sampled vectors
plt.scatter(vec[:, 0], vec[:, 1])
plt.xlabel('x-axis')
plt.ylabel('y-asis')
plt.title('Sample 100 uniformly distributed random vectors from $[-1, 1]^2$')
plt.show()
```

### Task 2.2
**<font>For each of the 100 vectors determine the minimum angle to all other vectors. Then compute the average of these minimum angles. Note that for two vectors $x, y$ the cosine of the angle between the two vectors is defined as </font>**

$$
\cos (\angle(x, y))=\frac{\langle x, y\rangle}{\|x\|\|y\|}
$$

```python
# define the average minimum angle against the dimension function
def avg_min_Angle(vec):
    n, d= vec.shape
    
    # The product of the multiplication of two vectors: <x, y>numerator
    p_vec = np.matmul(vec, vec.T) #(100,2)(2,100) (100,100)

    # The length of each vector
    len = np.linalg.norm(vec,axis=1) # (100,)

    # The product of the multiplication of length of each two vectors: ||x||||y||denominator
    p_len = np.dot(np.expand_dims(len, axis=1), np.expand_dims(len, axis=0))  # (100,1)(1,100) (100,100)

    # compute angle with the equation
    divide = np.divide(p_vec, p_len)
    np.fill_diagonal(divide, 0)  # diagonal 1->0
    angle = np.arccos(divide)  # (100, 100)

    # compute the average of these minimum angles
    avg_min_angle = np.mean(np.min(angle, axis=1)) 
    
    return avg_min_angle

avg_min_angle = avg_min_Angle(vec)
print('The avrage of these minimum angles is {}'.format(avg_min_angle))
```

### Task 2.3
**<font>Repeat the above for dimensions $d=1, \ldots, 1000$ and use the results to plot the average minimum angle against the dimension.</font>** 

```python
# calculate average minimum angles for 1~1000 dimension
def angle_dims(n, d):
    avg_min = np.zeros((d, 1))
    for i in range(1, d+1):  # i beginn from 1!
        vec = samples(n, i)
        avg_min[i-1,0] = avg_min_Angle(vec)
        
    return avg_min

avg_min_angle = angle_dims(100, 1000)

# plot average minimum angle
def show(dims, avg_min_angle):
    plt.plot(dims, avg_min_angle)
    plt.xlabel('Dimensions')
    plt.ylabel('average minimum angle')
    plt.title('average minimum angle of vectors in different dimensions')
    plt.grid()
    plt.show()

# dimension from 1 to 1000    
dims = range(1, 1000+1)
show(dims, avg_min_angle) 
```

### Task 2.4
**<font>Give an interpretation of the result. What conclusions can you draw for 2 randomly sampled vectors in a $d$-dimensional space?</font>**

```python
avg_min_angle = angle_dims(2, 1000)
dims = range(1, 1000+1)
show(dims, avg_min_angle) # dimension from 1 to 1000
```

As shown in the figure, we can draw a conclusion:

As the dimension increases, the average minimum angle between 2 randomly sampled vectors decreases. Besides, when the dimension tends to positive infinity, the curve converges to 1.5, that is, the angle between the two vectors tends to 90 degrees. On the contrary, the lower the dimension, the smaller the angle, and the smaller the vector distance.

$$
p(x, y)= \begin{cases}1 & \text { for } 0 \leq x \leq 1,0 \leq y \leq \frac{1}{2} \\ 0 & \text { otherwise }\end{cases}
$$

**<font>a joint density function for two random variables? </font>**

$$
\begin{align*}

\end{align*}
$$

Consequently, it is not a density function for two random variables, because the integral of pdf is not equal to 1.

### Task 4.1
**<font> [3 Points] Show that the covariance matrix $\mathbf{C}$ of any random variable $X \in \mathbb{R}^{p}$ is symmetric positive semidefinite, i.e. $\mathbf{C}=\mathbf{C}^{\top}$ and $\mathbf{x}^{\top} \mathbf{C x} \geq 0$ for any covariance matrix $\mathbf{C} \in \mathbb{R}^{p \times p}$ and any $\mathbf{x} \in \mathbb{R}^{p}$. </font>**

First we prove that $\mathbf{C}=\mathbf{C}^T$, assume a common covariance matrix $\mathbf{C}\in\mathbb{R}^{p\times p}$, expected value $\mathbf{\mu}=\mathbb{E}(\mathbf{X})$, $\mathbf{\mu}^T=\mathbb{E}(\mathbf{X}^T)$

$$
\begin{align}
\because \mathbf{C} &= \mathbb{E}((\mathbf{X}-\mathbf{\mu})(\mathbf{X}-\mathbf{\mu})^T) \\
              &= \mathbb{E}(\mathbf{XX}^T-\mathbf{X\mu}^T-\mathbf{\mu X}^T+\mathbf{\mu\mu}^T) \\
              &= \mathbb{E}(\mathbf{XX}^T)-\mathbb{E}(\mathbf{X\mu}^T)-\mathbb{E}(\mathbf{\mu X}^T)+\mathbb{E}(\mathbf{\mu\mu}^T) \\
              &= \mathbb{E}(\mathbf{XX}^T)-\mathbb{E}(\mathbf{X})\mathbf{\mu}^T-\mathbf{\mu}\mathbb{E}(\mathbf{X}^T)+\mathbf{\mu\mu}^T\mathbb{E}(1) \\
              &= \mathbb{E}(\mathbf{XX}^T)-\mathbf{\mu\mu}^T-\mathbf{\mu\mu}^T+\mathbf{\mu\mu}^T \\
              &= \mathbb{E}(\mathbf{XX}^T)-\mathbf{\mu\mu}^T \\
\therefore\mathbf{C}&= \mathbb{E}((\mathbf{X}\mathbf{X}^T)-\mathbf{\mu}\mathbf{\mu}^T\\ \hspace{3.83cm}
\end{align}
$$
Then we transpose the matrix $\mathbf{C}$
$$
\begin{align}
\because \mathbf{C}^T &= ({\mathbb{E}((\mathbf{X}\mathbf{X}^T)-\mathbf{\mu}\mathbf{\mu}^T})^T \\
               &= \mathbb{E}(\mathbf{XX}^T)^T-(\mathbf{\mu\mu}^T)^T \\
               &= \mathbb{E}((\mathbf{XX}^T)^T)-(\mathbf{\mu\mu}^T)^T \\
               &= \mathbb{E}((\mathbf{X}^T)^T(\mathbf{X})^T)-(\mathbf{\mu}^T)^T(\mathbf{\mu})^T \\
               &= \mathbb{E}(\mathbf{X}\mathbf{X}^T)-\mathbf{\mu}\mathbf{\mu}^T \\ \hspace{0.9cm}
\therefore\,\; \mathbf{C} &= \mathbf{C}^T \\ 
\end{align}
$$
From the above formulas, we can get $\mathbf{C}^T = \mathbf{C}$. Finally we prove that $\mathbf{x}^T\mathbf{Cx}\geq 0$

$$
\begin{align}
\because \mathbf{C} &= {\mathbb{E}((\mathbf{X}\mathbf{X}^T)-\mathbf{\mu}\mathbf{\mu}^T} \\
\Rightarrow \mathbf{x}^T\mathbf{C}\mathbf{x}&= \mathbf{x}^T\mathbb{E}((\mathbf{X}-\mathbf{\mu})(\mathbf{X}-\mathbf{\mu})^T)\mathbf{x} \\
&= \mathbb{E}(\mathbf{x}^T(\mathbf{X}-\mathbf{\mu})(\mathbf{X}-\mathbf{\mu})^T\mathbf{x}) \\
&= \mathbb{E}((\mathbf{x}^T(\mathbf{X}-\mathbf{\mu}))(\mathbf{x}^T(\mathbf{X}-\mathbf{\mu}))^T) \\
&= \mathbb{E}[(\mathbf{x}^T(\mathbf{X}-\mathbf{\mu}))^2] \\
\therefore \;\,\, & \mathbf{x}^T\mathbf{Cx} \geq 0
\end{align}
$$

Hence, $\mathbf{x}^T\mathbf{Cx} \geq 0$.
The covariance matrix $\mathbf{C}$ of any random variable $X \in \mathbb{R}^{p}$ is symmetric positive semidefinite.
